```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import numpy as np
actual = np.array(['dog', 'dog', 'dog', 'not dog', 'dog', 'not dog', 'dog', 'dog', 'not dog', 'not dog'])
predict = np.array(['dog', 'not dog', 'dog', 'not dog', 'dog', 'dog', 'dog', 'dog', 'not dog', 'not dog'])


w = confusion_matrix(actual, predict)

labels = ['dog', 'not dog']

sns.heatmap(w,
            annot=True,
            fmt='g',
            cmap='Oranges',
            xticklabels=labels,
            yticklabels=labels)

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import numpy as np

# 1. FIXED DATA
actual = np.array(['dog', 'dog', 'dog', 'not dog', 'dog', 'not dog', 'dog', 'dog', 'not dog', 'not dog'])
predict = np.array(['dog', 'not dog', 'dog', 'not dog', 'dog', 'dog', 'dog', 'dog', 'not dog', 'not dog'])

# 2. Generate the matrix (It will be 2x2)
w = confusion_matrix(actual, predict)

# 3. Create the Heatmap
labels = ['dog', 'not dog']

sns.heatmap(w,
            annot=True,
            fmt='g',
            cmap='Blues',
            xticklabels=labels,
            yticklabels=labels)

# --- ADDED TITLE HERE ---
plt.title('Confusion Matrix')

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --- STEP 1: PREPARE DATA ---
# (Fixed the typo: Removed the extra space in 'not dog')
actual = np.array(['dog', 'dog', 'dog', 'not dog', 'dog', 'not dog', 'dog', 'dog', 'not dog', 'not dog'])
predict = np.array(['dog', 'not dog', 'dog', 'not dog', 'dog', 'dog', 'dog', 'dog', 'not dog', 'not dog'])

# --- STEP 2: GENERATE TEXT REPORT ---
print("-- Classification Report ---")
print(classification_report(actual, predict))

# --- STEP 3: GENERATE VISUAL HEATMAP ---
# Calculate the confusion matrix
w = confusion_matrix(actual, predict)

# Define labels for the axes
labels = ['dog', 'not dog']

# Set up the figure size
plt.figure(figsize=(6, 5))

# Create the heatmap
sns.heatmap(w,
            annot=True,      # Show the numbers inside the boxes
            fmt='g',         # Use simple integers (no scientific notation)
            cmap='Blues',    # Color theme
            xticklabels=labels,
            yticklabels=labels)

# Add titles and labels
plt.title('Confusion Matrix')
plt.xlabel('Predicted Value')
plt.ylabel('Actual Value')

# Show the plot
plt.show()
```
